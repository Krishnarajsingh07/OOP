# 📘 Types of Classes in Python

Python me mainly ye types ki classes hoti hain:

---

## 1️⃣ Built-in Classes

Ye Python ke andar pehle se defined hoti hain.

### Example:
```python
l = [1, 2, 3]     # list class
s = "Hello"       # str class
n = 10            # int class
```

Examples:
- `int`
- `float`
- `str`
- `list`
- `tuple`
- `dict`

---

## 2️⃣ User-Defined Classes

Jo class programmer khud banata hai.

### Example:
```python
class Car:
    def __init__(self, brand):
        self.brand = brand

c1 = Car("BMW")
```

Yaha `Car` ek user-defined class hai.

---

## 3️⃣ Abstract Class

Ye class ka object directly create nahi kar sakte.
Iska use inheritance ke liye hota hai.

Abstract class banane ke liye `abc` module use hota hai.

### Example:
```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

---

## 4️⃣ Single Inheritance Class

Ek class sirf ek parent class se inherit karti hai.

```python
class Animal:
    pass

class Dog(Animal):
    pass
```

---

## 5️⃣ Multiple Inheritance Class

Ek class multiple parent classes se inherit karti hai.

```python
class Father:
    pass

class Mother:
    pass

class Child(Father, Mother):
    pass
```

    def __init__(self, brand, model, price):

# 📌 Function

**Definition:**

A function is a reusable block of code that performs a specific task.  
It is defined using the `def` keyword and can be called directly by its name.

**Example:**

```python
def greet():
    print("Hello")

greet()
```

---

# 📌 Method

**Definition:**

A method is a function that is defined inside a class and is associated with an object of that class.  
It is called using the object and usually takes `self` as its first parameter.

**Example:**

```python
class Student:

    def greet(self):
        print("Hello Student")

obj = Student()
obj.greet()
```

````python
L = [1,2,3]
len(L)  # Lenght is a function bcos it is outside if the class
L.append(4)  # Method is a function which is inside the class
````

# 📘 Meaning of `+` and `-` in UML Class Diagram

In UML Class Diagrams, symbols before attributes and methods
represent **access modifiers (visibility)**.

## 🔹 `+` (Plus Sign)

`+` means **Public**

- Can be accessed from anywhere (inside or outside the class).
- Example:
  
  + menu()
  
  This method can be called from outside the class.

---

## 🔹 `-` (Minus Sign)

`-` means **Private**

- Can be accessed only inside the class.
- Cannot be directly accessed from outside.
- Example:
  
  - pin : str
  
  This attribute is meant to be used internally inside the class.

---

## 🔹 Extra (For Knowledge)

- `#` → Protected (accessible in child classes)
- `~` → Package/Internal access

---

## ✅ Simple Meaning

`+` → Public (sab jagah se access ho sakta hai)  
`-` → Private (sirf class ke andar access ho sakta hai)
